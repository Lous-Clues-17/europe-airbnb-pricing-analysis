```python
import os
import pandas as pd

# listing files in data folder
print("Looking in 'data/':", os.listdir('data'))
```

```python
# loading one dataset, Amsterdam Weekdays
df = pd.read_csv('data/amsterdam_weekdays.csv')
```

```python
# preview  first 5 rows
display(df.head())
```

```python
# check the dimensions (Rows, Columns)
print(f"Dataset shape: {df.shape}")
```

```python
#check data types and mising values
df.info()
```

```python
# exploring individual column
df['room_type'].value_counts()
```

```python
# looking for pricing outliers. realSum is total price of listing
df[['realSum', 'dist', 'metro_dist']].describe()
```

```python
# checking for nulls in each column
print(df.isna().sum())
```

```python
# checking for duplicates
df.duplicated().sum()
```

```python
# filtering only listings with guest ratings over 90
high_rated_df = df[df['guest_satisfaction_overall'] > 90 ]
high_rated_df.head()
```

```python
#renaming realSum, multi, and biz columns
df = df.rename(columns={
    'realSum': 'total_listing_price', 
    'multi': 'is_multi_listing', 
    'biz': 'is_business_listing' 
    })

#verifying name change
print(df.columns)
```

```python
#dropping column Unnamed: 0 if it exists, if not, no error is raised
df = df.drop(columns=['Unnamed: 0'], errors='ignore')

#verifying drop
df.head()
```

```python
#sorting most expensive listings first
sorted_df = df.sort_values(by='total_listing_price', ascending=False)
sorted_df.head()
```

```python
# checking data types
print(df.dtypes)

#if 'is_multi_listing' or 'is_business_listing' are not boolean, convert:
df['is_multi_listing'] = df['is_multi_listing'].astype(bool)
df['is_business_listing'] = df['is_business_listing'].astype(bool)

# confirming data type changes
print(df.types)
```

```python
# summary: grouping by room type to see average price and satisfaction
# of weekday Airbnb bookings in Amsterdam

summary = df.groupby('room_type')[['total_listing_price', 'guest_satisfaction_overall']].mean()
print(summary)
```
